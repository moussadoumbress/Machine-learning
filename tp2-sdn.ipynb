```python
iris = datasets.load_iris()
X=iris.data
m,n=X.shape
classe=iris.target
labels=iris.target_names
colors=['C0','C1','C2']
fig, axs=plt.subplots(n,n)
for i in range(n):
    for j in range(n):
        for k in range(m):
            axs[i][j].scatter(X[k][i],X[k][j],color=colors[classe[k]],label=labels[classe[k]])

```

La visualisation (3,4) nous semble être la meilleure parce qu'elle sépare mieux les classes pour pouvoir faire éventuellemnt une classification des données.

```python
from sklearn.decomposition import PCA 
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

pca=PCA(n_components=2)
IrisPCA=pca.fit(X).transform(X)

lda=LinearDiscriminantAnalysis(n_components=2)
IrisLDA=lda.fit(X,classe).transform(X)

plt.figure()
for i in range(3):
    plt.scatter(IrisPCA[classe==i,0],IrisPCA[classe==i,1],color=colors[i],label=labels[i])

plt.title('PCA des données Iris')

plt.figure()

for i in range(3):
    plt.scatter(IrisLDA[classe==i,0],IrisLDA[classe==i,1],color=colors[i],label=labels[i])

plt.title('LDA des données Iris')
```

Il ne semble y avoir aucune différence fondamentale entre les deux visualisations. Elles paraissent symétreiques l'une de l'autre:la répartie de l'information au niveau des données que met en valeur le PCA se transfère de façon symétrique au niveau des classes (LDA).
