A**. First linear regression**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_csv("/home/ing/Bureau/tp_AP/tabBats.txt",sep=' ')
x=df["BOW"]
y=df["BRW"]
plt.figure()
plt.scatter(x,y)
plt.xlabel('BOW')
plt.ylabel('BRW')
plt.show()
```

we have a scatter plot. It seems to describe a line.

regression model equation:y=ax+b epsilon while epsilon gaussian random variable.

```python
from scipy import stats
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
#x=x.to_numpy()
x=x.reshape(-1,1)
regr = linear_model.LinearRegression()
regr.fit(x,y)
print(regr.coef_)
print(regr.intercept_)
print(regr.score(x,y))
y_predit=regr.predict(x)
plt.scatter(x,y)

plt.plot(x,y_predit,color="blue")
plt.xlabel('BOW')
plt.ylabel('BRW')
plt.title('1st linear regression with an outlier point')
plt.show()
```

R²=95% means that 95% of information of BRO is contained in BOW.
we have the line regression which predict the brain weigth in function of the body weight.
the regression is not weel done because of an outlier point.

```python
corr=np.corrcoef(df["BOW"],y)

print(corr)
residual=y_predit-y

error_m=np.square(residual)/len(y)

plt.scatter(y_predit,error_m)
plt.xlabel('y_predict')
plt.ylabel('RSS')
plt.show()
```

BOW and BRW are positively and strongly correlated. The RSS is not quite near of zero. the model can be improved by removing the outlier.

Pour la seconde regression linéaire, on a supprimé la valeur aberrante et la droite passe presque par tous les points.
